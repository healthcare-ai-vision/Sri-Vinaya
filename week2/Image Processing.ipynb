```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
# Load sample images from skimage
img1 = data.camera()      # grayscale image
img2 = data.astronaut()   # color image

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(img1, cmap='gray')
plt.title("Image 1 - Camera")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img2)
plt.title("Image 2 - Astronaut")
plt.axis('off')
```

```python
#Grayscale Conversion
gray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)

plt.imshow(gray2, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
```

```python
#Histogram Visualization
plt.hist(gray2.ravel(), bins=256)
plt.title("Histogram of Grayscale Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
```

```python
#Image Resizing
resized = cv2.resize(gray2, (128, 128))

plt.imshow(resized, cmap='gray')
plt.title("Resized Image (128x128)")
plt.axis('off')
```

```python
#Edge Detection
edges = cv2.Canny(resized, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis('off')
```

```python
#Noise Additiona nd Removal
# Add Gaussian noise
noise = np.random.normal(0, 20, resized.shape)
noisy_image = resized + noise
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

# Remove noise using Gaussian Blur
denoised = cv2.GaussianBlur(noisy_image, (5,5), 0)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.imshow(resized, cmap='gray')
plt.title("Original")

plt.subplot(1,3,2)
plt.imshow(noisy_image, cmap='gray')
plt.title("Noisy")

plt.subplot(1,3,3)
plt.imshow(denoised, cmap='gray')
plt.title("Denoised")

plt.axis('off')
```

Dataset 1: NIH Chest X-ray Dataset

Type: X-ray images

Number of images: ~112,000

Classes: 14 disease labels (Pneumonia, Effusion, Cardiomegaly, etc.)

Dataset imbalance: Yes (many normal images, fewer disease cases)

Challenges: Label noise, multi-label classification, image quality variation

Summary:
This dataset is widely used for detecting chest diseases using CNNs. It helps in early diagnosis and reducing radiologist workload.

Dataset 2: ISIC Skin Lesion Dataset

Type: Dermoscopic skin images

Number of images: ~25,000

Classes: Melanoma, nevus, benign keratosis, etc.

Dataset imbalance: Severe class imbalance

Challenges: Similar-looking lesions, lighting variations, artifacts

Summary:
The ISIC dataset supports skin cancer detection tasks. CNN models learn color and texture patterns critical for melanoma classification.
